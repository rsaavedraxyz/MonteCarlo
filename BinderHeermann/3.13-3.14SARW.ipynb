# Self-avoiding random walk

```julia
using Random
using PyPlot
include("PlotSingleWalk.jl")
```

```julia
mutable struct Pos
    x::Float64
    y::Float64
end

function Base.:+(a::Pos, b::Pos)
    return Pos(a.x + b.x, a.y + b.y)
end;
```

```julia
function generate_one_step()
    moves = [Pos(1.0,0.0), Pos(-1.0,0.0), Pos(0.0,1.0), Pos(0.0,-1.0)]
    rn::Int64 = ceil(4*rand())
    return moves[rn]
end;
```

```julia
generate_one_step()
```

```julia
function single_SARW(nsteps::Int64=10)
    """ Generate a single self-avoiding random walk
    nsteps = number of steps of the random walk
    """
    traj = [Pos(0.0,0.0)]
    lattice = zeros(Int64, 2*nsteps + 1, 2*nsteps + 1)
    lattice[nsteps+1, nsteps+1] = 1
    r = Pos(0.0,0.0)
    step = 1
    while true
        Δ = generate_one_step()
        li::Int64 = 1 + nsteps + Int(r.x)
        lj::Int64 = 1 + nsteps + Int(r.y)
        linew::Int64 = 1 + nsteps + Int(r.x+Δ.x)
        ljnew::Int64 = 1 + nsteps + Int(r.y+Δ.y)
        trapped::Bool = lattice[li-1,lj] == 1 && lattice[li+1,lj] == 1 && lattice[li,lj-1] == 1 && lattice[li,lj+1] == 1
        if step == nsteps
            break
        elseif trapped
            break
        elseif lattice[linew, ljnew] != 1
            lattice[linew, ljnew] = 1
            r.x += Δ.x
            r.y += Δ.y
            step += 1
            push!(traj, Pos(r.x, r.y))
        end
    end
    println("Terminated at step = $step")
    return traj
end;
```

```julia
nsteps = 100
r_vec = single_SARW(nsteps)
x_vec = [r.x for r in r_vec]
y_vec = [r.y for r in r_vec]
PlotSingleWalk.plot_single_walk(x_vec, y_vec);
```

## Sampling the SARW

```julia
function sample_SARW(nsamples::Int64=100, nsteps::Int64=100)
    final_steps = [0 for i ∈ 1:nsamples]
    r2_end = Pos(0.0,0.0)
    r4_end = Pos(0.0,0.0)
    for sample ∈ 1:nsamples
        step = 0
        lattice = zeros(Int64, 2*nsteps + 1, 2*nsteps + 1)
        lattice[nsteps+1, nsteps+1] = 1
        r = Pos(0.0,0.0)
        while true
            Δ = generate_one_step()
            li::Int64 = 1 + nsteps + Int(r.x)
            lj::Int64 = 1 + nsteps + Int(r.y)
            linew::Int64 = 1 + nsteps + Int(r.x+Δ.x)
            ljnew::Int64 = 1 + nsteps + Int(r.y+Δ.y)
            trapped::Bool = lattice[li-1,lj] == 1 && lattice[li+1,lj] == 1 && lattice[li,lj-1] == 1 && lattice[li,lj+1] == 1
            if step == nsteps
                break
            elseif trapped
                break
            elseif lattice[linew, ljnew] != 1
                lattice[linew, ljnew] = 1
                r.x += Δ.x
                r.y += Δ.y
                step += 1
            end
        end
        final_steps[sample] = step
        r2_end += Pos(r.x^2, r.y^2)
        r4_end += Pos(r.x^4, r.y^4)
    end
    r2_end = Pos(r2_end.x/nsamples, r2_end.y/nsamples)
    r4_end = Pos(r4_end.x/nsamples, r4_end.y/nsamples)
    return final_steps, r2_end, r4_end
end
```

```julia
nsamples = 2000
Nvec = [i for i ∈ 10:200]
nwalks_with_N = [0.0 for i ∈ Nvec]
r2vec = [Pos(0.0,0.0) for i ∈ Nvec]
r4vec = [Pos(0.0,0.0) for i ∈ Nvec]
xerr_vec = [0.0 for i ∈ Nvec]
yerr_vec = [0.0 for i ∈ Nvec]
for (i,Ni) ∈ enumerate(Nvec)
    final_steps, r2_end, r4_end = sample_SARW(nsamples, Ni)
    nwalks_with_N[i] = length(filter(j->(j==Ni), final_steps))
    r2vec[i] = r2_end
    r4vec[i] = r4_end
    xerr_vec[i] = r4_end.x - r2_end.x^2
    yerr_vec[i] = r4_end.y - r2_end.y^2
end
```

```julia
PyPlot.rc("figure", figsize = (8, 3))

subplot(121)
W = nwalks_with_N ./ nsamples
plot(Nvec, W)

xlim(0,last(Nvec))
ylim(0,1.0)
xlabel("N")
ylabel("W")

subplot(122)
S = log.(nwalks_with_N ./ nsamples)
plot(Nvec, S)

xlim(0,last(Nvec))
#ylim(0,1.0)
xlabel("N")
ylabel(L"$\ln(W)$")

tight_layout()
```

```julia
PyPlot.rc("figure", figsize = (8, 3))

subplot(121)
x2vec = [r.x for r in r2vec]
y2vec = [r.y for r in r2vec]
plot(Nvec, x2vec, label=L"$\langle X^2\rangle$")
plot(Nvec, y2vec, label=L"$\langle Y^2\rangle$")

xlim(0,last(Nvec))
xlabel("N")
legend()

subplot(122)
rsq_xerr_vec = sqrt.(xerr_vec) ./ x2vec
rsq_yerr_vec = sqrt.(yerr_vec) ./ y2vec
plot(Nvec, rsq_xerr_vec, label=L"$X_{error}$")
plot(Nvec, rsq_yerr_vec, label=L"$Y_{error}$")

xlim(0,last(Nvec))
xlabel("N")
legend()

tight_layout()
```
