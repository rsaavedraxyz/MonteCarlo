# Random walk using a BAD random number generator

Here we cover exercises 3.1-3.2 from the book on Monte Carlo Simulations in Statistical Physics by K. Binder and D. W. Heerman.

The basic idea to generate a random number is a folding mechanism, e.g.

$$ x_i = ax_{i-1} + c \mod{m} $$

where $m, a, c, x_0$ are integers with $m>x_0$ and $a,c,x_0>0$.

We define the following (pseudo-)random number generator

```julia
function modulo_RNG(arr_size::Int64=1; seed::Int64=3)
    """ Generate an array of random numbers from 0 to 1 
    arr_size = size of the array
    seed = seed of the RNG
    """
    mult::Int64 = 1277
    modulo::Int64 = 2^17
    rmodulo::Float64 = modulo
    random = zeros(Float64, arr_size)
    for i âˆˆ 1:arr_size
        seed = seed * mult
        seed = seed % modulo
        random[i] = seed / rmodulo
    end
    return random
end;
```

```julia
modulo_RNG(10, seed=5)
```

Now we will use `modulo_RNG` to generate a single random walk in two dimensions. First we define a function that returns a vector for each spatial coordinate

```julia
function single_2DRW(ran_vec::Vector{Float64})
    """ Generate a single 2D random walk
    ran_vec = vector of random numbers
    """
    x = 0
    y = 0
    x_vec = zeros(Int64, nsteps+1)
    y_vec = zeros(Int64, nsteps+1)
    for (i,ran_f) in enumerate(ran_vec)
        ran_i::Int64 = ceil(4*ran_f)
        if ran_i == 1
            x += 1
        elseif ran_i == 2
            y += 1
        elseif ran_i == 3
            x -= 1
        else
            y -= 1
        end
        x_vec[i+1] = x
        y_vec[i+1] = y
    end
    return x_vec, y_vec
end;
```

We plot the random walk using the `PyPlot` library

```julia
using PyPlot
PyPlot.rc("figure", figsize = (4, 4));

nsteps = 20
ran_vec = rand(nsteps)
x_vec, y_vec = single_2DRW(ran_vec)

# get arrows
xx_vec = [xi + (xf - xi)/2 for (xi, xf) in zip(x_vec[1:1:end], x_vec[2:1:end])]
yy_vec = [yi + (yf - yi)/2 for (yi, yf) in zip(y_vec[1:1:end], y_vec[2:1:end])]
vx_vec = [xf - xi for (xi, xf) in zip(x_vec[1:1:end], x_vec[2:1:end])]
vy_vec = [yf - yi for (yi, yf) in zip(y_vec[1:1:end], y_vec[2:1:end])]

# plot
plot(x_vec, y_vec, color="b", lw=3)
quiver(xx_vec, yy_vec, vx_vec, vy_vec,
    color="b", scale_units="xy", scale=2, pivot="mid", 
    headwidth=10, headlength=10, headaxislength=5)
scatter(first(x_vec), first(y_vec), s=60, marker="o", color="b")
scatter(last(x_vec), last(y_vec), s=60, marker="s", color="b")

# formatting
xlo, xhi = minimum(x_vec), maximum(x_vec)
ylo, yhi = minimum(y_vec), maximum(y_vec)
xlabel("X")
ylabel("Y")
vlines(xlo:xhi, ylo, yhi, color="k", alpha=0.2)
hlines(ylo:yhi, xlo, xhi, color="k", alpha=0.2)
xticks([]), yticks([])
axis("equal")
axis("off");
```

Note the resulting walk has some "dead ends" where the walk has reversed its direction. Simple random walks are not useful for modelling polymers (chains of colloids), as in that case one expects the colloids not to overlap. Thus for polymers we require self-avoiding random walks.

## Sampling random walks

Let us examine the results on the end-to-end distance of random walks generated with the algorithm from above.

We perform several random walks and record the average end-to-end distance of each component $\langle X^2\rangle$ and $\langle Y^2\rangle$, as well as the fluctuations $\langle X^4\rangle$ and $\langle Y^4\rangle$.

```julia
function sample_2DRW(nsamples::Int64=10, nsteps::Int64=10)
    """ Generate samples of 2D random walks
    nsamples = number of random walk samples
    nsteps = number of steps of each random walk
    """
    x2av = 0.0
    y2av = 0.0
    x4av = 0.0
    y4av = 0.0
    ran_vec = modulo_RNG(nsamples*nsteps)
    for i âˆˆ 1:nsamples
        xi = 0.0
        yi = 0.0
        for j âˆˆ 1:nsteps
            rn::Int64 = ceil(4*ran_vec[nsteps*(i-1)+j])
            if rn == 1
                xi += 1.0
            elseif rn == 2
                yi += 1.0
            elseif rn == 3
                xi -= 1.0
            else
                yi -= 1.0
            end
        end
        xsq = xi^2
        ysq = yi^2
        x2av += xsq
        y2av += ysq
        x4av += xsq*xsq
        y4av += ysq*ysq
    end
    x2av /= nsamples
    y2av /= nsamples
    x4av /= nsamples
    y4av /= nsamples
    av = (x2=x2av, y2=y2av, x4=x4av, y4=y4av)
    return av
end;
```

We define arrays to store the simulation data

```julia
nsamples = 2000
nsteps_vec = 0:30:3000

x2_vec = [0.0 for i in nsteps_vec]
y2_vec = [0.0 for i in nsteps_vec]
x4_vec = [0.0 for i in nsteps_vec]
y4_vec = [0.0 for i in nsteps_vec]
xerr_vec = [0.0 for i in nsteps_vec]
yerr_vec = [0.0 for i in nsteps_vec]
etime_vec = [0.0 for i in nsteps_vec]
for (i,nsteps) in enumerate(nsteps_vec)
    stats = @timed begin
        av = sample_2DRW(nsamples, nsteps)
    end
    x2_vec[i] = av.x2
    y2_vec[i] = av.y2
    x4_vec[i] = av.x4
    y4_vec[i] = av.y4
    xerr_vec[i] = av.x4 - av.x2.^2
    yerr_vec[i] = av.y4 - av.y2.^2
    etime_vec[i] = stats.time
end
```

Now we plot

```julia
PyPlot.rc("figure", figsize = (8, 3))

subplot(121)
plot(nsteps_vec, x2_vec, label=L"$\langle X^2\rangle$")
plot(nsteps_vec, y2_vec, label=L"$\langle Y^2\rangle$")
plot(nsteps_vec, 0.5*nsteps_vec, linestyle=":", color="k", label="N/2")

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();

subplot(122)
plot(nsteps_vec, x2_vec .+ y2_vec, label=L"$\langle R^2\rangle$")
plot(nsteps_vec, nsteps_vec, linestyle=":", color="k", label="N")

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();
```

We observe each direction of the walk is equally probable. We also know that the results must be $\langle R^2\rangle\sim t$, but from the numerical treatment we have to conclude a different law. What went wrong?

Let us look at the error of the end-to-end distances

```julia
rsq_xerr_vec = sqrt.(xerr_vec) ./ x2_vec
rsq_yerr_vec = sqrt.(yerr_vec) ./ y2_vec

PyPlot.rc("figure", figsize = (4, 3))

plot(nsteps_vec, rsq_xerr_vec)
plot(nsteps_vec, rsq_yerr_vec)

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0.0,2.0)
xlabel("steps")
ylabel("error");
```

The error is not constant!

```julia
PyPlot.rc("figure", figsize = (4, 3))

plot(nsteps_vec, etime_vec)

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
ylabel("execution time (seconds)");
```

## Testing the quality of the RNG

The quality of a random number generator can be seen if one tries to completely fill a $d$-dimensional lattice.

We can map the interval (0,1) to the discrete set $\{0,1,2,\ldots\}$ of integer numbers. This mapping allows us to construct random vectors $(x_1,\ldots,x_d)$ of integers in a $d-$dimensional hypercube. Each site reached with a random vector derived from the generator is marked. A good generator must cover the entire hypercube.

```julia
function plot_RNG_correlations(ran_vec::Vector{Float64})
    """ Plot correlations between random numbers
    ran_vec = vector of random numbers (0,1)
    """
    if isodd(length(ran_vec))
        throw("Wrong ran_vec dimensions! its length should be an even number")
    end
    len = Int(length(ran_vec)/2)
    x = 2*len*ran_vec[1:len]
    y = 2*len*ran_vec[len+1:2*len]
    scatter(x,y, s=4, alpha=0.5)
    xticks([])
    yticks([])
end;
```

```julia
PyPlot.rc("figure", figsize=(7,4))

# GOOD region of the RNG
subplot(121)
ran_vec_good = modulo_RNG(1500, seed=3)
plot_RNG_correlations(ran_vec_good)
ax = gca()
ax.set_xlim([0,1500])
ax.set_ylim([0,1500])

# BAD region of the RNG
subplot(122)
ran_vec_bad = modulo_RNG(5000, seed=3)
plot_RNG_correlations(ran_vec_bad)
ax = gca()
ax.set_xlim([0,1500])
ax.set_ylim([0,1500]);
```

Clearly the generator performs badly, i.e., shows duplet correlations. Only parts of the lattice are visited and many sites are not visited at all.
