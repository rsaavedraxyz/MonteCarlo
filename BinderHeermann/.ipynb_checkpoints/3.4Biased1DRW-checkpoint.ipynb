# Biased random walk

```julia
using Random
using PyPlot
```

```julia
function samples_Biased1DRW(nsamples::Int64=10, nsteps::Int64=10, bias::Float64=0.5)
    """ Generate sample biased random walks in 1D
    nsamples = number of samples
    nsteps = step length of each random walk
    """
    x2av = 0.0
    Random.seed!(3)
    ran_vec = rand(nsamples*nsteps)
    for i ∈ 1:nsamples
        xi = 0.0
        for j ∈ 1:nsteps
            rn::Float64 = ran_vec[nsteps*(i-1)+j]
            if rn < bias
                xi += 1.0
            else
                xi -= 1.0
            end
        end
        xsq = xi^2
        x2av += xsq
    end
    x2av /= nsamples
    return x2av
end;
```

```julia
nsamples = 100
nsteps_vec = 0:5:2700
x2avA_vec = [0.0 for i in nsteps_vec]
x2avB_vec = [0.0 for i in nsteps_vec]
x2avC_vec = [0.0 for i in nsteps_vec]
for (i,nsteps) in enumerate(nsteps_vec)
    x2avA_vec[i] = samples_Biased1DRW(nsamples, nsteps, 0.49)
    x2avB_vec[i] = samples_Biased1DRW(nsamples, nsteps, 0.50)
    x2avC_vec[i] = samples_Biased1DRW(nsamples, nsteps, 0.51)
end
```

```julia
PyPlot.rc("figure", figsize = (4, 3))
plot(nsteps_vec, x2avA_vec, label="p=0.45")
plot(nsteps_vec, x2avB_vec, label="p=0.5")
plot(nsteps_vec, x2avC_vec, label="p=0.55")
xlim(first(nsteps_vec), last(nsteps_vec))
xlabel("steps")
ylabel(L"$\langle X^2\rangle$")
legend()
```
