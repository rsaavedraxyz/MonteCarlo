# Nonreversal random walk (ignorant approach)

In a nonreversal random walk each step has to be a “forward” step. By “forward” step we mean that the walker is not allowed to make an immediate reversal. It is, however, allowed to cross its own path or at some later time retrace some of the steps made earlier.

For the numerical treatment of the nonreversal random walk (NRRW) by simple sampling we can follow at least two possible routes:

1. Ignore at each step that the walker came from somewhere (except at the origin). If the step that the walker takes goes back to the one it came from, another try is made until a “forward” step is found. Here we choose at each step one of the four sites in our two-dimensional problem with probability 1/4.

2. The other possibility is to remember where we came from and choose only from the three possible neighboring sites. Here the probability is taken to be 1/3 for each of the three possible directions. There are always three directions even if all the sites the walker can go to have been visited before at some time. Only the one visited immediately before the current step is prohibited.

For now we take the ignorant approach and may knock at closed doors several times.

```julia
function single_NRRW(nsteps::Int64=10)
    """ Generate a single nonreversal random walk
    nsteps = number of steps of the random walk
    """
    x, y = 0, 0
    xtemp, ytemp = 0, 0
    xprev, yprev = 0, 0
    x_vec, y_vec = zeros(Int64, nsteps+1), zeros(Int64, nsteps+1)
    for i ∈ 1:nsteps
        while true
            xtemp, ytemp = x, y #save the current position
            rn::Int64 = ceil(4*rand())
            if rn == 1
                x += 1
            elseif rn == 2
                y += 1
            elseif rn == 3
                x -= 1
            else
                y -= 1
            end
            if xprev != x || yprev != y #nonreversal step: exit loop
                break
            else  #reversal step: go back to previous position
                x, y = xtemp, ytemp
            end
        end
        xprev, yprev = xtemp, ytemp
        x_vec[i+1], y_vec[i+1] = x, y
    end
    return x_vec, y_vec
end;
```

```julia
include("RandomWalk.jl")
using .RandomWalk:  plot_single_walk, sample_2DRW
```

```julia
nsteps = 20
x_vec, y_vec = single_NRRW(nsteps)

plot_single_walk(x_vec, y_vec);
```

## Sampling nonreversal random walks

We perform several random walks and record the average end-to-end distance of each component $\langle X^2\rangle$ and $\langle Y^2\rangle$, as well as the fluctuations $\langle X^4\rangle$ and $\langle Y^4\rangle$.

```julia
function sample_NRRW(nsamples::Int64=10, nsteps::Int64=10)
    """ Sample nonreversal random walks in two-dimensions
    nsamples = number of samples
    nsteps = number of steps of each random walk
    
    Returns a named tuple of averages (x2=<x²>, y2=<y²>, x4=<x⁴>, y4=<y⁴>)
    """
    x2_av, y2_av, x4_av, y4_av = 0.0, 0.0, 0.0, 0.0
    ran_vec = rand(nsamples*nsteps)
    for i ∈ 1:nsamples
        xtemp, ytemp = 0.0, 0.0
        xprev, yprev = 0.0, 0.0
        xi, yi = 0.0, 0.0
        for j ∈ 1:nsteps
            while 1 == 1
                xtemp, ytemp = xi, yi
                rn::Int64 = ceil(4*rand())
                if rn == 1
                    xi += 1.0
                elseif rn == 2
                    yi += 1.0
                elseif rn == 3
                    xi -= 1.0
                else
                    yi -= 1.0
                end
                if xprev != xi || yprev != yi
                    break
                else
                    xi, yi = xtemp, ytemp
                end
            end
            xprev, yprev = xtemp, ytemp
        end
        xsq, ysq = xi^2, yi^2
        x2_av += xsq
        y2_av += ysq
        x4_av += xsq*xsq
        y4_av += ysq*ysq
    end
    x2_av /= nsamples
    y2_av /= nsamples
    x4_av /= nsamples
    y4_av /= nsamples
    return x2_av, y2_av, x4_av, y4_av
end;
```

We employ arrays to record the samples

```julia
nsamples = 1000
nsteps_vec = 0:10:2700

x2_vec = [0.0 for i in nsteps_vec]
y2_vec = [0.0 for i in nsteps_vec]
x4_vec = [0.0 for i in nsteps_vec]
y4_vec = [0.0 for i in nsteps_vec]
etime_vec = [0.0 for i in nsteps_vec]
for (i,nsteps) in enumerate(nsteps_vec)
    stats = @timed begin
        x2_vec[i], y2_vec[i], x4_vec[i], y4_vec[i] = sample_NRRW(nsamples, nsteps)
    end
    etime_vec[i] = stats.time
end;
```

We plot the end-to-end distance

```julia
using PyPlot
PyPlot.rc("figure", figsize = (8, 3))

subplot(121)
plot(nsteps_vec, x2_vec, label=L"$\langle X^2\rangle$")
plot(nsteps_vec, y2_vec, label=L"$\langle Y^2\rangle$")
plot(nsteps_vec, nsteps_vec, linestyle="--", color="k", label=L"$N$")

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();

subplot(122)
plot(nsteps_vec, x2_vec .+ y2_vec, label=L"$\langle R^2\rangle$")
plot(nsteps_vec, 2*nsteps_vec, linestyle="--", color="k", label=L"$2N$")

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();
```

Plot the fluctuations

```julia
PyPlot.rc("figure", figsize = (8, 3))

subplot(121)
plot(nsteps_vec, x4_vec, label=L"$\langle X^4\rangle$")
plot(nsteps_vec, y4_vec, label=L"$\langle Y^4\rangle$")
plot(nsteps_vec, 3*nsteps_vec.^2, linestyle="--", color="k", label=L"$3N^2$") # is this correct?

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();

subplot(122)
plot(nsteps_vec, x4_vec .+ y4_vec, label=L"$\langle R^4\rangle$")
plot(nsteps_vec, 6*nsteps_vec.^2, linestyle="--", color="k", label=L"$6N^2$") # is this correct?

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
legend();
```

The execution time

```julia
PyPlot.rc("figure", figsize = (4, 3))

plot(nsteps_vec, etime_vec)

xlim(first(nsteps_vec), last(nsteps_vec))
ylim(0)
xlabel("steps")
ylabel("execution time (seconds)");
```
